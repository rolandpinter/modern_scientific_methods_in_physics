```python
# Some Python imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in logfile
df = pd.read_csv("logfile", sep = ",", header = 0)
df.head()
```

```python
plt.figure(figsize = (17,7))

plt.subplot(1,2,1)
plt.plot(df["N"], df["avg_exec_mean_seq"], "bo-", label = "Sequential")
plt.plot(df["N"], df["avg_exec_mean_par"], "go-", label = "Parallel")

plt.title("Mean computation", fontsize = 20)
plt.xlabel("N (number of data)", fontsize = 20)
plt.ylabel(r'Average execution time [$\mu s$]', fontsize = 20)
plt.legend(fontsize = 20)
plt.yscale("log")
plt.xscale("log")
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)

plt.subplot(1,2,2)
plt.plot(df["N"], df["avg_exec_dev_seq"], "bo-", label = "Sequential")
plt.plot(df["N"], df["avg_exec_dev_par"], "go-", label = "Parallel")

plt.title("Deviation computation", fontsize = 20)
plt.xlabel("N (number of data)", fontsize = 20)
plt.ylabel(r'Average execution time [$\mu s$]', fontsize = 20)
plt.legend(fontsize = 20)
plt.yscale("log")
plt.xscale("log")
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)

plt.show()
```
